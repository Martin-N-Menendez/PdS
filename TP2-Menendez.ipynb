```python
#%% Inclusion de librerias
import numpy as np
# If the code is in a file called plot_zplane.py
from plot_zplane import zplane 
from scipy import signal
import matplotlib.pyplot as plt

#%% Declaracion de variables para N = 3
numerador = np.array([1/3, 1/3, 1/3])
denominador = np.array([1,0,0])
```

```python
#%% Diagrama de polos para N = 3
zplane(numerador,denominador)
```

```python
#%% Calculo de modulo y fase
# W = Frecuencia 0 a Pi # H = Respuesta en frecuencia
w, h = signal.freqz(numerador,denominador)              # Frec y rta
modulo = abs(h)                                         # Modulo de la rta
angulos = np.unwrap(np.angle(h))                        # Fase de la rta
```

```python
#%% Ploteo de resultados
fig, ax1 = plt.subplots()                               # Subplots
ax1.plot(w, modulo, 'b')                                # Plot de modulo
ax1.set_ylabel('Amplitud [normalizado]', color='b')     # Label Y

ax2 = ax1.twinx()                                       # Sgte subplot
ax2.plot(w, angulos, 'r')                               # Plot de fase
ax2.set_ylabel('Fase [radianes]', color='r')            # Label Y

ax1.set_title('Resultados para N = 3')                  # Titulo
ax1.set_xlabel('Frequencia [rad/muestra]')              # Label X
ax2.grid()                                              # Grilla
ax2.axis('tight')                                       # Grosor de linea
plt.show()                                              # Mostrar grafico
```

```python
#%% Inclusion de librerias
import numpy as np
# If the code is in a file called plot_zplane.py
from plot_zplane import zplane 

from scipy import signal
import matplotlib.pyplot as plt
from  matplotlib import patches
from matplotlib.figure import Figure
from matplotlib import rcParams
import scipy.io as sio
import scipy.fftpack as sc
```

```python
mat = sio.loadmat('ECG_TP4.mat')

ecg_lead = mat['ecg_lead']  
qrs_pattern1 = mat['qrs_pattern1']
heartbeat_pattern1 = mat['heartbeat_pattern1']
heartbeat_pattern2 = mat['heartbeat_pattern2']
qrs_detections = mat['qrs_detections']
```

```python
#%% Configuracion de sampleo
fs = 1000                                      #1KHz
ts = 1/fs

ecg_N = len(ecg_lead)
N = len(ecg_lead[:ecg_N])
tt = np.linspace(0.0, (N-1)*ts, N)
```

```python
#%% Ploteo de resultados
fig, ax1 = plt.subplots()                               # Subplots
ax1.plot(tt, ecg_lead[:ecg_N],'b')                              # Plot de ECG
ax1.set_ylabel('Amplitud [normalizado]', color='b')     # Label Y

ax1.set_title('Registro ECG')                  # Titulo
ax1.set_xlabel('Tiempo [Segundos]')              # Label X
ax1.grid()                                               # Grilla
ax1.axis('tight')                                       # Grosor de linea
plt.show()                                              # Mostrar grafico
```

```python
ff = np.linspace(0.0, fs/2, N//2)

spectrum = np.abs(sc.fft(ecg_lead[:ecg_N], axis=0))
spectrum = (1/max(spectrum))*spectrum
halfSpectrum = spectrum[:N//2] 
```

```python
#%% Ploteo de resultados
fig, ax1 = plt.subplots()                               # Subplots
ax1.plot(ff, halfSpectrum,'r')                             # Plot de FFT de ECG
ax1.set_ylabel('Amplitud [normalizado]', color='r')     # Label Y

ax1.set_title('FFT: Registro ECG')                  # Titulo
ax1.set_xlabel('Frecuencia [Hz]')              # Label X
ax1.grid()                                               # Grilla
ax1.axis('tight')                                       # Grosor de linea
plt.show()                                              # Mostrar grafico
```
